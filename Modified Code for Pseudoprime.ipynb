```python
#GENERAL 
# Determine the order of k mod n
def ordd(k,n):
    if gcd(k,n)>1:
        o=0
    else: 
        j=k
        o=1
        while j!=1:
            j=(k*j)%n
            o=o+1
    return(o)        


#check if n a pseudoprime to base k
def is_psp(k,n):
    if gcd(k,n)>1:
        return(1==0)
    else:
        return((n-1)%ordd(k,n)==0)


#get a factor of k^n-1 between m and M
def a_factor_g(k,n,m,M):
    r=m
    ans="None"
    while r<M:
        if gcd(k,r)==1:
            if n%ordd(k,r)==0:
                ans=r
                r=M+1
            else:
                r=r+1
        else:
            r=r+1
    return(ans)        


#Algo input n=pq, find p and q
def Divgen_g(k,n):
    r=a_factor_g(k,n,3,50000)
    d=gcd(r-1,n)
    if d==n:
        r=a_factor_g(k,n,r+1,50000)
        d=gcd(r-1,n)
    while not is_psp(k,d):
        Divgen_g(k,d)
    return(d)

#matrix
def ps_matrix(k,L):
    M=[L]
    for p in L:
        S=[]
        for q in L:
            if (p-1)%ordd(k,q)==0 and (q-1)%ordd(k,p)==0:
                S=S+[1]
            else:
                S=S+[0]
        M=M+[S]
    return(matrix(M))
#new algo
def ALGO_g(k,n):
    S=[]
    p=Divgen_g(k,n)
    S=S+[p]
    n=n/p
    while not is_psp(k,n):
        p=Divgen_g(k,n)
        S=S+[p]
        n=n/p
    S=S+[n]    
    return(ps_matrix(k,S))    
```
